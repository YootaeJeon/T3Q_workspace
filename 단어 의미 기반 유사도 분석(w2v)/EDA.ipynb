형태소 분석을 통해 특저 품사의 단어만 저장(일반명사, 고유명사)

```python
from kiwipiepy import Kiwi       # https://github.com/bab2min/Kiwi
```

```python
class CustomTokenizer:
    def __init__(self, tagger):
        self.tagger = tagger
    def __call__(self, sent):
        sent = sent[:1000000]
        word_tokens = self.tagger.tokenize(sent) # 형태소 출력
        ''' NNG 일반명사, NNP 고유 명사'''
        word_tokens = [token.form for token in word_tokens if token.tag in  ['NNG', 'NNP','SL']]                      # Token(form='오늘', tag='NNG', start=0, len=2), Token(form='은', tag='JX', start=2, len=1),
        result = [word for word in word_tokens if len(word) > 1]
        return result
     
```

```python
import pandas as pd

# 파일 경로

#file_path = 'english_data_with_filenames.txt'
file_path = './english_data.txt'

# 파일을 열어 각 줄을 읽고 처리합니다
with open(file_path, 'r', encoding='utf-8') as file:
    data = []
    for line in file:
        # ':' 기호를 기준으로 파일 이름과 내용을 분리
        name, content = line.split(':', 1)
        data.append({'name': name.strip(), 'content': content.strip()})

# 데이터 프레임 생성
df = pd.DataFrame(data)
df
```

```python
df = df.drop([7,60,74])
df= df.reset_index(drop=True)
```

```python
import re
line = df['content']
# 한글과 띄어쓰기, 영어를 제외한 모든 글자
compile = re.compile("[^ \+|a-z|A-Z|ㄱ-ㅣ가-힣]+")
for i in range(len(line)):
    a = compile.sub("",line[i])
    line[i] = a
    
print(line[0])
```

```python
def remove_stopwords_from_list(content_list, stopwords):
    """Remove stopwords"""
    filtered_list = []

    for sentence in content_list:
        filtered_sentence = " ".join([word for word in sentence if word not in stopwords])
        filtered_list.append(filtered_sentence)
        
    return filtered_list
```
